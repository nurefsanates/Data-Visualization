```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
veri=pd.read_csv("olimpiyatlar_temizlenmis.csv")
veri.head()
```

```python
#SAÇILIM GRAFİĞİ:iki farklı değişkenin arasındaki ilişkiyi berlilemek için kullanılır
def sacilimGrafikCiz():
    sns.scatterplot(x='boy',y='kilo',data=veri)
    plt.title('Boy ve Ağırlık İlişkisi')
    plt.show()
sacilimGrafikCiz()
```

```python
sns.set_style("darkgrid")
sacilimGrafikCiz()
```

```python
sns.set_style("whitegrid")
sacilimGrafikCiz()
```

```python
sns.scatterplot(x='boy',y='kilo',hue='cinsiyet', style='madalya', size='yas',sizes=(15,200),data=veri)
plt.title('Madalyaya göre boy ve ağırlık dağılımı')
plt.show()
```

```python
#DOĞRUSAL REGRESYON
sns.regplot(x='boy',y='kilo',data=veri,marker='+',scatter_kws={'alpha':0.2})
plt.title('Boy ve Ağırlık Dağılımı')
plt.show()
```

```python
#RENK PALETLERİ
sns.scatterplot(x='boy',y='kilo',hue='madalya',data=veri,palette='Set1')
plt.title('Madalyaya göre Boy ve Ağırlık Dağılımı')
plt.show()
```

```python
sns.scatterplot(x='boy',y='kilo',hue='madalya',data=veri,palette='rocket')
plt.title('Madalyaya göre Boy ve Ağırlık Dağılımı')
plt.show()
```

```python
#ÇİZGİ GRAFİĞİ
sns.lineplot(x='boy',y='kilo',data=veri)
plt.title('Boy ve Kilo')
plt.show()
```

```python
#KATEGORİK ÇİZGİ GRAFİĞİ
sns.lineplot(x='boy',y='kilo',hue='cinsiyet',data=veri)
plt.title('Cinsiyete göre Boy ve Ağırlık')
plt.show()
```

```python
#HİSTOGRAM
sns.displot(veri,x='kilo')
plt.ylabel('Frekans')
plt.title('Ağırlık Histogramı')
plt.show()
```

```python
#KATEGORİK HİSTOGRAM
sns.displot(veri,x='kilo',hue='cinsiyet')
plt.ylabel('Frekans')
plt.title('Cinsiyete Göre Ağırlık Histogramı')
plt.show()
```

```python
sns.displot(veri,x='kilo',col='cinsiyet',multiple='dodge')
plt.show()
```

```python
#İKİ BOYUTLU HİSTOGRAM
sns.displot(veri,x='kilo',y='boy',kind='kde')
plt.xlabel('Ağırlık')
plt.ylabel('Boy')
plt.title('Ağırlık-Boy Histogramı')
plt.show()
```

```python
sns.displot(veri,x='kilo',y='boy',hue='cinsiyet',kind='kde')
plt.xlabel('Ağırlık')
plt.ylabel('Boy')
plt.title('Cinsiyete Göre Ağırlık-Boy Histogramı')
plt.show()
```

```python
sns.kdeplot(data=veri,x='kilo',y='boy',hue='cinsiyet',fill=True)
plt.xlabel('Ağırlık')
plt.ylabel('Boy')
plt.title('Cinsiyete Göre Ağırlık-Boy Histogramı')
plt.show()
```

```python
#ÇUBUK GRAFİĞİ
sns.barplot(x='madalya',y='boy',data=veri)
plt.title('Madalyalara Göre Boy Grafikleri')
plt.show()
```

```python
sns.barplot(x='madalya',y='boy',data=veri,hue='cinsiyet')
plt.title('Madalyalara Göre Boy Grafikleri')
plt.show()
```

```python
#Catplot ile kategorik şekilde oluşturabiliriz
sns.catplot(x='madalya',y='yas',hue='cinsiyet',col='sezon',data=veri,kind='bar',height=4,aspect=0.7)
plt.show()
```

```python
#KUTU GRAFİĞİ
sns.boxplot(x='sezon',y='boy',data=veri)
plt.show()
```

```python
sns.boxplot(x='sezon',y='boy',hue='cinsiyet',data=veri,palette='Set3')
plt.show()
```

```python
veri_gecici= veri.loc[:,["yas","kilo","boy"]]
sns.boxplot(data=veri_gecici,orient='h',palette='Set2')
plt.show()
```

```python
sns.catplot(x='sezon',y='boy',hue='cinsiyet',col='madalya',data=veri,kind='box',height=4,aspect=0.7)
```

```python
#ISI HARİTASI
veri_yedek=veri.loc[:,["yas","kilo","boy","yil"]]
sns.heatmap(veri_yedek.corr(),annot=True,linewidths=0.5,fmt='.3f')
plt.show()
```

```python
#KEMAN GRAFİĞİ
sns.violinplot(x='sezon',y='boy',data=veri)
plt.show()
```

```python
sns.violinplot(x='sezon',y='boy',data=veri,hue='cinsiyet')
plt.show()
```

```python
#split ile yarıya bölüp genel şekli daha az şekille görebiliriz
sns.violinplot(x='sezon',y='boy',data=veri,hue='cinsiyet',split=True)
plt.show()
```

```python
sns.catplot(x='sezon',y='boy',hue='cinsiyet',col='madalya',data=veri,kind='violin',split=True,height=4,aspect=0.7)
```

```python
#ORTAKLIK GRAFİĞİ
sns.jointplot(data=veri,x='kilo',y='boy',hue='sezon',kind='kde')
plt.show()
```

```python
g=sns.JointGrid(data=veri,x='kilo',y='boy')
g.plot_joint(sns.histplot)
g.plot_marginals(sns.boxplot)
plt.show()
```

```python
#EŞ GRAFİĞİ
sns.pairplot(veri)
plt.show()
```

```python
#aynı tabloları tekrar tekrar çizmemek için
g=sns.PairGrid(veri)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot,fill=True)
g.map_diag(sns.histplot,kde=True)
```

```python
#SAYMA GRAFİĞİ
sns.countplot(x='sehir',data=veri)
plt.xticks(rotation=90) #yazı derecesi
plt.show()
```
